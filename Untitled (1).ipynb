# ● The mean sales generated by each zone.

```python
MEAN_A=df['Zone - A'].mean()
MEAN_B=df['Zone - B'].mean()
MEAN_C=df['Zone - C'].mean()
MEAN_D=df['Zone - D'].mean()
```

```python
df1=pd.DataFrame({'Zone-A':[MEAN_A],'Zone-B':[MEAN_B],'Zone-C':[MEAN_C],'Zone-D':[MEAN_D]})
```

```python
df1
```

```python
df1.plot(kind='bar',title='MEAN SALES GENERATED BY EACH ZONE')
```

Zone D has the highest mean and zone A has the lowest

# ● Total sales generated by all the zones for each month.

# ● Check whether all the zones generate the same amount of sales.

Setting Hypothesis

H0: All zones generates the same amount of sales

H1: All zones generate different amount of sales

```python
from scipy.stats import f_oneway
```

```python
f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

Since pvalue is less than alpha, we will reject the null hypothesis.Hence the conclusion is all zones generated different amount of sales
